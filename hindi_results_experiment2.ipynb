Since it's very difficult to translate each word into English we will be annotating all the neighbours into these three categories:   
SW -> Same Word(obviously with different sense)  
SC -> Same Category  
PDC -> Partially Different Category  
TDC -> Totally Different Category   
While annotating these words we used this [dictionary](http://www.cfilt.iitb.ac.in/hindishabdamitra/) provided by [IIT Bomabay](http://www.iitb.ac.in/) from where we also took the data for these experiments and this categorization is skewed by personal knowledge and considerations, others might have different opinions about some words.
